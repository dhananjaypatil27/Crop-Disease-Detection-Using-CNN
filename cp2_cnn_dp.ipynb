```python
# Convert the multi-class PlantVillage dataset into a binary (healthy vs non_healthy) dataset by copying images into respective folders.
import os, shutil

SOURCE_DIR = "/content/PlantVillage/PlantVillage" # original PlantVillage
BASE_DIR   = "/content/PlantVillage_split"        # binary dataset

# create base folders
os.makedirs(os.path.join(BASE_DIR, "healthy"), exist_ok=True)
os.makedirs(os.path.join(BASE_DIR, "non_healthy"), exist_ok=True)

# populate them
for folder in os.listdir(SOURCE_DIR):
    folder_path = os.path.join(SOURCE_DIR, folder)
    if not os.path.isdir(folder_path):
        continue

    target = "healthy" if "healthy" in folder.lower() else "non_healthy"

    for img in os.listdir(folder_path):
        src = os.path.join(folder_path, img)
        dst = os.path.join(BASE_DIR, target, img)
        if not os.path.exists(dst):
            shutil.copy(src, dst)

print("--> healthy & non_healthy folders created")
```

```python
 # Split each class dataset into (70% Train, 15% Val, 15% Test) and copy images into respective folders.

for cls in classes:
    cls_path = os.path.join(BASE_DIR, cls)

    images = os.listdir(cls_path)
    random.shuffle(images)

    total = len(images)
    train_end = int(0.7 * total)
    val_end   = int(0.85 * total)

    train_imgs = images[:train_end]
    val_imgs   = images[train_end:val_end]
    test_imgs  = images[val_end:]

    for img in train_imgs:
        shutil.copy(
            os.path.join(cls_path, img),
            os.path.join(BASE_DIR, "train", cls, img)
        )

    for img in val_imgs:
        shutil.copy(
            os.path.join(cls_path, img),
            os.path.join(BASE_DIR, "val", cls, img)
        )

    for img in test_imgs:
        shutil.copy(
            os.path.join(cls_path, img),
            os.path.join(BASE_DIR, "test", cls, img)
        )

print("--> Train / Val / Test split completed")
```

```python
# Verify the number of images in each train, validation, and test folder for both classes.

for split in ["train", "val", "test"]:
    for cls in classes:
        path = os.path.join(BASE_DIR, split, cls)
        print(split, cls, "в†’", len(os.listdir(path)))
```

```python
# Create augmented training data generator and normal validation/test generators using MobileNetV2 preprocessing,
# then load images from train/val/test folders for binary classification.

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

# Train generator (with augmentation)
train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,  # REQUIRED for MobileNetV2
    rotation_range=30,
    zoom_range=0.2,
    horizontal_flip=True
)

# Validation & Test generator (NO augmentation)
val_test_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_data = train_gen.flow_from_directory(
    BASE_DIR + "/train",
    target_size=(128,128),
    batch_size=64,
    class_mode="binary"
)

val_data = val_test_gen.flow_from_directory(
    BASE_DIR + "/val",
    target_size=(128,128),
    batch_size=64,
    class_mode="binary"
)

test_data = val_test_gen.flow_from_directory(
    BASE_DIR + "/test",
    target_size=(128,128),
    batch_size=64,
    class_mode="binary",
    shuffle=False
)
```

```python
# Build a MobileNetV2-based transfer learning model by freezing the pretrained base and adding custom layers for binary classification.

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

base_model = MobileNetV2(
    input_shape=(128,128,3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False   # рџ”‘ key for speed

x = GlobalAveragePooling2D()(base_model.output)
output = Dense(1, activation="sigmoid")(x)

model = Model(base_model.input, output)
```

```python
# Compile the model with Adam optimizer and binary crossentropy loss, then train it for 15 epochs using training and validation data.

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=15
)
```

```python
# Evaluate the trained model on the test dataset and print the final test accuracy.

test_loss, test_acc = model.evaluate(test_data)
print("Test Accuracy:", test_acc)
```

```python
# Generate predictions on test data and compute confusion matrix and classification report to evaluate model performance in detail.

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

y_true = test_data.classes
y_pred = (model.predict(test_data) > 0.5).astype(int).ravel()

print(confusion_matrix(y_true, y_pred))
print(classification_report(y_true, y_pred, target_names=["Healthy", "Diseased"]))
```

```python
# Plot training and validation accuracy and loss curves to visualize model performance over epochs.

import matplotlib.pyplot as plt

plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend()
plt.title("Accuracy")
plt.show()

plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Loss")
plt.show()
```
